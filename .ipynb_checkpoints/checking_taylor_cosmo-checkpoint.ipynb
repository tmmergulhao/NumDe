# Importing libraries

```python
import numpy as np
import numerical_derivative as nd
import generating_grid
import itertools
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.style.use('ggplot')
import time

kvec = generating_grid.kvec
```

# Loading the grid and computing the derivatives

NumDe can be used to Taylor expand any function. For this example, in particular, I Taylor expanded the outputs of CLASS-PT with respect to As, omega_cdm and h. Firstly, I ran the code generating_grid.py. It generated an output file containing the CLASS-PT output evaluated in a grid of cosmological parameters. Here, I will load it and test its precision with the exact output from CLASS-PT.

```python
ND_cosmo = nd.FiniteDerivative()
ND_cosmo.load_grid_from_file("output/PK_grid.npz")
```

The derivatives were already computed. In case you only have the grid, you just have to run

```python
ND_cosmo.take_derivative_ALL(save_file="output/PK_grid.npz")
```

The details of the grid is in one of its methods:

```python
print("Grid setup:", ND_cosmo.expansion_table)
print("Grid size:", ND_cosmo.N_grid)
```

The expansion is around A = 2.083, omega_cdm = 0.12 and h = 0.6736. Let us see how it performs if we evaluate the function far from this point

# Testing the expansion

```python
A = 4
omega_cdm = 0.08
h = 0.56

operators_computed = generating_grid.ComputePk([A, omega_cdm, h])
```

```python
list_to_test_single = [1, 30, 31, 32, 33, 4,5,7,8,14,26,34,35,36,37,9,28,29,38,39]
for this_operator in list_to_test_single:
    plt.figure(figsize = (3,3))
    plt.title(this_operator)
    operators_taylor = ND_cosmo.TaylorExpand([A, omega_cdm, h], 4, stencil=4)
    plt.plot(kvec, 1-operators_taylor[this_operator]/operators_computed[this_operator])
    plt.ylim((-0.05,0.05))
    plt.xlim((0,0.2))
    plt.show()
    plt.close()
```

With a Taylor expansion up to 4th order and stencil = 4 you can have a precision of roughly 1% for each PT operator. Notice that after we combine all these operators to have the final galaxy power spectrum the final error will be smaller.
